```python
import base64
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders

def send_email_with_pdf(service, to, subject, body, pdf_path):
    message = MIMEMultipart()
    message['to'] = to
    message['subject'] = "Data Analyst Available – Python, Power BI, SQL | Daksh Sankhla"
    
    # Email body
    message.attach(MIMEText(body, 'plain'))
    
    # Attach PDF
    with open(pdf_path, 'rb') as f:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload(f.read())
    
    encoders.encode_base64(part)
    part.add_header('Content-Disposition', f'attachment; filename="/home/question/Downloads/Daksh Professional CV.pdf"')
    message.attach(part)
    
    raw = base64.urlsafe_b64encode(message.as_bytes()).decode()
    service.users().messages().send(userId='me', body={'raw': raw}).execute()

# Usage
send_email_with_pdf(service, 'petrifiedlightning.spacejunk@gmail.com', 'Hello', 
                    'See attached!', '/home/question/Downloads/Daksh Professional CV.pdf')
```
