Finding house rent using beautiful soup

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib as plt

url = "https://www.magicbricks.com/flats-for-rent-in-gurgaon-price-20000-to-30000-pppfr"
response = requests.get(url)
soup = BeautifulSoup(response.content,"html.parser")

houses = soup.find_all("div",class_ ="mb-srp__card")


titles = []
for house in houses:
    title = house.find("h2",class_="mb-srp__card--title").text.replace(" ","")
    titles.append(title)

prices = []
for house in houses:
    price = house.find("div",class_="mb-srp__card__price--amount").text
    prices.append(price)

other_costs = []
for house in houses:
    cost = house.find("div",class_="mb-srp__card__price--other-charges").text
    other_costs.append(price)

tenant_pref = []
for house in houses:
    tenant = house.find("div",class_="mb-srp__card__summary--value").text
    tenant_pref.append(tenant)

description = []
for house in houses:
    desc = house.find("div",class_="mb-srp__card--desc--text").text
    description.append(desc)

appartment_names = []
for house in houses:
    appartment = house.find("div",class_="mb-srp__card__society")
    if appartment is not None:
        appartment_names.append(appartment.text)

card_details = soup.find_all("div", class_="mb-srp__card__summary--value")
card_details
plain_text = []
for detail in card_details:
    plain_text.append(detail.get_text())


areas = []
for text in plain_text:
    if "sqft" in text:
        areas.append(text)
floor = []
for text in plain_text:
    if "out of" in text:
        floor.append(text)

list_of_lists = [ appartment_names, prices, other_costs, areas, floor, titles, description, tenant_pref]
df = pd.DataFrame (list_of_lists)
dnew = df.transpose()
columns_list = ["appartment_names", "prices", "other_costs", "area", "floor", "bhk", "description", "tenant_pref"]
dnew.columns = columns_list
dnew.head()
```
